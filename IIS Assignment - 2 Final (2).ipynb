```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split as tt_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.preprocessing import StandardScaler,Normalizer
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_excel("Data for IISLab Assignment.xlsx")
df.head()
```

```python
df=df.drop(['ID','DOB','10board','12board','12graduation'],axis='columns')
df
```

```python
df=pd.get_dummies(data=df,columns=['Gender','Degree','Specialization','CollegeState'])
```

```python
# df = df.sample(frac=1).reset_index(drop=True)
```

```python
training=df.drop('High-Salary',axis='columns')
testing=df['High-Salary']
```

```python
train_x,test_x,train_y,test_y=tt_split(training,testing,train_size=0.70)
```

```python
sc=StandardScaler()
train_x=sc.fit_transform(train_x)
test_x=sc.transform(test_x)
```

```python
# nm=Normalizer().fit(train_x)
# nm.transform(train_x)
# nm.transform(test_x)
```

```python
# train_x.std(axis=0)
```

```python
max_iter=10000
```

```python
lr=LogisticRegression(C=11.288378916846883, class_weight='balanced', penalty='l1',
                   random_state=0, solver='liblinear')
```

```python
lr.fit(train_x,train_y)
```

```python
# param_grid = [    
#     {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
#     'C' : np.logspace(-4, 4, 20),
#     'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
#     'max_iter' : [100, 1000,2500, 5000]
#     }
# ]
# clf = GridSearchCV(lr, param_grid = param_grid, cv = 3, verbose=True, n_jobs=-1)
# best_clf = clf.fit(train_x,train_y)
# best_clf.best_estimator_
# print (f'Accuracy - : {best_clf.score(test_x,test_y):.3f}')
```

```python
predict=lr.predict(test_x)
```

```python
cm=confusion_matrix(predict,test_y)
```

```python
acc=accuracy_score(predict,test_y)
print("\nAccuracy of model is:",acc*100)
```

```python
print("Confusion matrix is:\n",cm)
```

```python
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
print("Normalized confusion matrix (indicating class-wise accuracies) is:")
cm.diagonal()
```

```python
print("\nClassification Report is:")
print(classification_report(predict,test_y,target_names=['High','Low']))
```
